```python
import os

# Tentukan path direktori yang berisi file-file yang ingin direname
directory = 'AmbilData8/'

# Iterasi melalui semua file dalam direktori
for i, filename in enumerate(os.listdir(directory), start=1):
    # Cek apakah filename adalah file (bukan direktori)
    if os.path.isfile(os.path.join(directory, filename)):
        # Buat nama baru dengan urutan sekuensial
        new_filename = '{:d}.jpg'.format(i)
        # Ubah nama file
        old_name = os.path.join(directory, filename)
        new_name = os.path.join(directory, new_filename)
        os.rename(old_name, new_name)
        print(f"File {old_name} berhasil diubah menjadi {new_name}.")
```

```python
import os

# Tentukan path direktori yang berisi file-file yang ingin direname
directory = 'AmbilData9/'

# Mendapatkan daftar file dalam direktori dan mengurutkannya berdasarkan angka dalam nama file
files = sorted(os.listdir(directory), key=lambda x: int(x.split('.')[0]))

# Iterasi melalui semua file dalam direktori
for i, filename in enumerate(files, start=1):
    # Cek apakah filename adalah file (bukan direktori)
    if os.path.isfile(os.path.join(directory, filename)):
        # Ubah nama file
        old_name = os.path.join(directory, filename)
        new_name = os.path.join(directory, '{:05d}.jpg'.format(i))
        os.rename(old_name, new_name)
        print(f"File {old_name} berhasil diubah menjadi {new_name}.")
```
